# 🎥Netflix Global Content Analysis

# 📖Introduction

  This project presents a comprehensive analysis of the **Netflix Movies and TV Shows** dataset. Using Python's data science ecosystem, the study explores patterns in content distribution, including genres, release trends, country-wise production, and content types. The objective is to gain insights into Netflix's content strategy and understand the factors that contribute to its position as a global leader in the streaming industry.

#**🚀Netflix Data Analysis Key Business Insights**

* 🎥**Content On Netflix**

  ***Movies*** dominate the platform which make up significantly larger portion of the Netflix catalog compared to ***TV Shows*** , indicating Netflix hosts more Movie Content than Series.




* **Top Countries Producing Netflix Content**

  The ***United States*** contributes to the highest amount of Netflix Content , followed by countries such as ***India*** and the ***United Kingdom*** which shows that Netflix Content is concentrated in a few major regions.

* **Rating Distribution On Netflix**

  Most of the Netflix Content falls under ratings such as **TV-MA** and **TV-14** , indicating the platform is predominantly designed for *Adults and Teenagers*.


* **Netflix Content Release Year Analysis**

  The number of Netflix releases significantly increased after 2000 , rapidly increased after 2015 , indicating rapid growth in Content Production and expansion of the platform during recent years.

* 🔝**Top Generes on Netflix**

  Generes like ***International Movies*** ,**Dramas**, ***Comedies*** are most popular in the dataset; showing that Netflix offers a strong variety of globally produced content.

* **Top Movies with longest Duration**

  The Movie '***Black Mirror: Bandersnatch***' has the ***longest duration*** amongst all the movies on Netflix followed by '***Headspace: Unwind Your Mind***' and '***The School of Mischief***'.

* **Top TV Shows with most number of seasons**

  The TV Show '***Grey's Anatomy***' has most number of Seasons on Netflix with overall 17 Seasons, followed by Shows '***NCIS***' and '***Supernatural***' with overall 15 Seasons.

* 📊**Monthly Distribution of Netflix Releases**

  The analysis of content release patterns reveals a clear seasonal trend in Netflix's publishing strategy. ***July*** emerges as the month with the highest number of releases , followed by ***December*** and ***September***.

* **🌍Content Preferences by Country**

  The analysis highlights distinct content consumption pattern across countries.The ***United States*** leads in both *movies* and *TV Shows* preferences, highlighting its central role in Netflix's content landscape. For ***Movies***, it is followed by ***India***, indicating a strong preference for film-based content. In Contrast, for ***TV Shows*** , the ***United Kingdom*** ranks second,suggesting a higher inclination toward episodic content.

# Key Libraries : Pandas , Numpy , Matplotlib , Seaborn and Plotly

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('/content/sample_data/netflix_titles.csv.zip')
```

```python
df.shape
```

```python
df.head()
```

#Handling NULL Values

```python
df.isnull().sum()
```

```python
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['date_added'].fillna('Unknown', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
df['rating'].fillna('Not Rated', inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

# Content On Netflix

```python
#Content on Netflix
df['type'].value_counts().plot(kind = 'pie',autopct='%.2f')
plt.title('Content on Netflix')
plt.show()
```

#Top 5 Countries producing Netflix Content

```python
#Top 5 Countries producing Netflix Content
plt.style.use('dark_background')
country = df[df['country'] != 'Unknown']['country'].value_counts().sort_values(ascending=False).head().plot(kind='bar')
for container in country.containers :
  country.bar_label(container)
plt.title('Top 5 Countries producing Netflix Content')
plt.xlabel('Countries')
plt.ylabel('Number of contents')
plt.show()
```

```python
df['rating'].unique()
```

#Distribution of Ratings on Netflix Content

```python
#Ratings on Netflix Content
plt.style.use('classic')
plt.figure(figsize=(12,7))
rating = df['rating'].value_counts().plot(kind='barh')
for container in rating.containers :
  rating.bar_label(container)
plt.title('Distribution of Ratings on Netflix')
plt.xlabel('Rating Count')
plt.ylabel('Rating')
plt.show()
```

#Netflix Releases Year Wise Distribution

```python
#Netflix Releases Year Wise Distribution
plt.style.use('dark_background')
plt.grid()
plt.plot(df['release_year'].value_counts().sort_index(),linewidth=3,color='b',alpha=1)
plt.title('Netflix Releases Year Wise Distribution')
plt.xlabel('Number of Netflix Releases')
plt.ylabel('Year')
plt.show()
```

Analyzing it further with different distributions plots

```python
#Let's analyze it further with further plots
plt.style.use('default')
sns.displot(df['release_year'] , kind = 'kde')
```

```python
plt.figure(figsize=(6,5))
data = df['release_year'].value_counts()
fig = px.line(data, title='Netflix Releases Year Wise Distribution',template='plotly_dark',markers=True)
fig.update_traces(line = dict(width=5),marker=dict(color='green'))
fig.show()
```

#Top 5 Countries prefering TV Shows

```python
#Top 5 Countries prefering TV Shows
plt.figure(figsize=(8,5))
pre = df[df['type'] == 'TV Show']
p = pre[df['country'] != 'Unknown']
pr = p.groupby('country')['type'].count().sort_values(ascending=False).head().reset_index()
plot = sns.barplot(data=pr,x='country',y='type',hue='country')
for i in plot.containers :
  plot.bar_label(i,label_type='center')
plt.title('Top 5 Countries prefering TV Shows')
plt.xlabel('Country')
plt.ylabel('Count of TV Shows')
plt.show()
```

#Top 5 Countries prefering Movies

```python
#Top 5 Countries prefering Movies
plt.figure(figsize=(8,5))
pre = df[df['type'] == 'Movie']
p = pre[df['country'] != 'Unknown']
pr = p.groupby('country')['type'].count().sort_values(ascending=False).head().reset_index()
plot = sns.barplot(data=pr,x='country',y='type',hue='country')
for j in plot.containers :
  plot.bar_label(j,label_type='center')
plt.title('Top 5 Countries prefering Movies')
plt.xlabel('Country')
plt.ylabel('Count of Movies')
plt.show()
```

```python
data = df.copy()
```

#Top 10 Popular Generes on Netflix

```python
#Top 10 Popular Generes on Netflix
data['listed_in'] = data['listed_in'].str.split(',')
data = data.explode('listed_in')
data['listed_in'] = data['listed_in'].str.strip()
top_genres = data['listed_in'].value_counts().sort_values(ascending=True)
plt.figure(figsize=(12,6))
genres = top_genres.tail(10).plot(kind='barh' , edgecolor='purple')
for val in genres.containers :
  plt.bar_label(val)
plt.title('Top 10 Popular Generes on Netflix')
plt.xlabel('Count of Genres')
plt.ylabel('Generes')
plt.show()
```

#Top 10 Movies with longest Duration

```python
df['duration'] = df['duration'].str.replace('min','')
df['duration'] = df['duration'].str.replace('Season','')
df['duration'] = df['duration'].str.replace('Seasons','')
df['duration'] = df['duration'].str.replace('s','')
dframe = df[df['duration'] != 'Unknown'].reset_index()
dframe['duration'] = dframe['duration'].astype(int)
```

```python
#Top 10 Movies with longest Duration
plt.figure(figsize=(8,5))
d = dframe[dframe['type'] == 'Movie']
longMov = d.groupby('title')['duration'].max().sort_values(ascending=False).head(10).reset_index()
movies = sns.barplot(data=longMov,y='title',x = 'duration',hue='title')
for k in movies.containers :
  movies.bar_label(k)
plt.title('Top 10 Movies with longest Duration')
plt.ylabel('Title of the Movie')
plt.xlabel('Duration in minutes')
plt.show()
```

#Top 10 TV Shows with Most Number of Seasons

```python
#Top 10 TV Shows with Most Number of Seasons
plt.figure(figsize=(8,5))
da = dframe[dframe['type'] == 'TV Show']
longShows = da.groupby('title')['duration'].max().sort_values(ascending=False).head(10).reset_index()
shows = sns.barplot(data=longShows,y='title',x = 'duration',hue='title')
for p in shows.containers :
  shows.bar_label(p)
plt.title('Top 10 TV Shows with Most Number of Seasons')
plt.ylabel('Title of the TV Show')
plt.xlabel('Number of Seasons')
plt.show()
```

#Months with most number of Netflix Releases

```python
#Months with most number of Netflix Releases
plt.style.use('dark_background')
date = df[df['date_added'] != 'Unknown']
date['date_added'] = date['date_added'].str.strip()
date['date_added'] = pd.to_datetime(date['date_added'],format='%B %d, %Y')
date['month'] = date['date_added'].dt.month_name()
plt.figure(figsize=(14,6))
plt.plot(date['month'].value_counts(),markerfacecolor='red',marker='o',linewidth = 2,color='c')
plt.show()
```

```python
months = date['month'].value_counts().sort_values(ascending=False).plot(kind='bar',color='c',edgecolor='white')
for p in months.containers :
  months.bar_label(p)
plt.show()
```

#Top 10 Directors on Netflix

```python
#Top 10 Directors on Netflix
direct = df[df['director'] != 'Unknown']
top_directors = direct['director'].value_counts().sort_values(ascending=True).tail(10).reset_index()
top_directors.columns = ['Director','Count of Content']
fig = px.bar(top_directors,x='Count of Content',y='Director',title='Top 10 Directors on Netflix',color='Count of Content',text='Count of Content')
fig.update_layout(template='plotly_dark')
fig.show()
```
